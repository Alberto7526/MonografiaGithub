```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import division_dataset as div_d
import altair as alt
```

```python
df = pd.read_csv('./Datasets/sales_train.csv')
df_shops_categories = div_d.transform_data_shop_category(df)
df_shops_categories.head()
```

```python
df_shops_categories.describe()
```

```python
df_shops_categories.info()
```

```python
sum_per_month = df_shops_categories.drop(['shop_id','item_category_id'],axis=1).sum()
sns.set()
fig, ax = plt.subplots(figsize=(15,4))
ax.scatter(range(0,34),sum_per_month,color='black')
ax.plot(range(0,34),sum_per_month,label = 'Sales per month',color='gold',linewidth = 2)
plt.xlabel('Months',fontsize=18)
plt.ylabel('Sales',fontsize=18)
plt.title('Total sales per month',fontsize=20)
plt.legend()
plt.savefig('Sales_per_month.png',dpi=300)
plt.show()
```

```python
df_shops = div_d.transform_data_shop(df)
df_shops = df_shops.drop(['shop_id'],axis=1)
df_shops.head()
```

```python
df_shops_trans = df_shops.transpose()
df_shops_trans.head()
```

```python
fig, ax = plt.subplots(10,6,figsize=(15,44))
sns.set()
num = 0
for k in range(0,10):
    for i in range (0,6):
        x = df_shops_trans[num]
        #ax.plot(x,label = 'shop_'+str(i))
        ax[k,i].bar(range(0,34),x,color='darkgoldenrod')
        ax[k,i].plot(range(0,34),x,color='black')
        ax[k,i].set_title('shop_'+str(num))
        num +=1
#plt.savefig('Sales_per_shop.png',dpi=100)
plt.show()
```
